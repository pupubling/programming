##Example 1

```
for loop
for (initialize; test; update) {
  ...
}
```

```
for ( 初始化 initialize ; 是否繼續執行的條件判斷 test ; 更新變數值 update ) { 
  ...

}
```

其中三個區域的每個區域所做的事情還可稍加變化。如果什麼都沒有，寫成
```
for ( ; ; ) {
  printf("Do something.\n");
}
```
作用則相當於無窮迴圈。


---

其他 Assignment 符號: `+=，-=，*=，/=，%=`

這些可看成是簡寫的 assignment 符號，`x += 5;`相當於 `x = x + 5;` 而 `y %= 7;` 相當於 `y = y % 7;` 以此類推。 

在 `for` 迴圈的第三部份 (更新變數值) 用這樣的寫法會比較簡潔，但你不一定要用這樣的寫法，

只要順著自己的習慣 就可以。知道有這樣的寫法，看到別人的程式用到這種寫法才能知道它的意思。

在 `for` 迴圈的第一部份 (初始化) 和第三部份 (更新變數值)，其實允許我們做兩個以上的 statements。

````python
%%writefile E04_01.c
#include <stdio.h>
int main(void)
{
	char ch;
	for (ch = 'a'; ch <= 'z'; ch++) {
		printf("The ASCII value for %c is %d.\n", ch, ch);
	}
	return 0; 
}
````

````python
%%shell
gcc E04_01.c -o E04_01
./E04_01
````

##Example 2
for loop 多做幾件事

`for (n=1, cost=FIRST_PACK; n<=10; n++, cost+=NEXT_PACK)`

````python
%%writefile E04_02.c
#include <stdio.h>
#define FIRST_PACK 7
#define NEXT_PACK 5
int main(void)
{
	int n, cost;
	printf(" packs costs\n");
	for (n=1, cost=FIRST_PACK; n<=10; n++, cost+=NEXT_PACK) {
		printf("%5d     $%-5d\n", n, cost);
	}
	return 0; 
}
````

````python
%%shell
gcc E04_02.c -o E04_02
./E04_02
````

##Example 4

使用 `getchar()` 和 `putchar()`

````python
%%writefile E04_04.c
/* 
使用者輸入一串字元(讀到 '\n' 就結束)
把這串字元裡全部的英文字母都轉成大寫再輸出 
*/
#include <stdio.h>
#include <ctype.h>
int main(void)
{
   char ch;

   while ( (ch=getchar()) != '\n' ) {         
      putchar(tolower(ch));
   }

   putchar('\n');   

   return 0;
}
````

````python
%%shell
gcc E04_04.c -o E04_04
./E04_04
````

##Example 5

`if` (條件判斷) {
}
列出因數


````python
%%writefile E04_05.c
/*
輸入一個正整數
程式必須列出除了 1 和該正整數本身以外的全部因數
因數必須從小到大排列
如果無法因數分解
則成是必須顯示這個正整數是質數
*/
#include <stdio.h>
int main(void)
{
   int num, div;
   int isPrime = 1;

   printf("Please enter an integer (q to quit): \n");

   scanf("%d", &num);
   for (div=2; div*div <= num ; div++ ) {
       if (num%div == 0) {
         printf("%d, ", div); 
         isPrime = 0;        
       }
   }
   for (div-=2; div > 1; div--) {
       if (num%div == 0) {
         printf("%d, ", num/div);         
       }
   }

   if (isPrime == 1) {
      printf("%d is prime.\n", num);
   } else {
      printf(" are divisors of %d\n", num);
   }
   return 0;
}
````

````python
%%shell
gcc E04_05.c -o E04_05
./E04_05
````

##Example 6

`if` expression

````python
%%writefile E04_06.c
/*
定義一個 function (取名為 check)
檢驗下面的 expression 
根據 expression 的值
輸出對應的資訊 "True" 或 "False"
*/
#include <stdio.h>
void check(int cond)
{
    if (cond) {
        printf("True\n");
    } else {
        printf("False\n");
    } 
}
int main(void)
{
   int x = 5; 
   int y = 3;
   
   check(x > 2 && y == 3);
   check( !(x < 2 || y == 3) );
   check(x != 0 && y);
   check(x == 0 || !y);
   check(x != y && (20/x) < y);
   check( !(y>5 || y<2) && !(x%5) );
   check(!x && x);
       
   return 0;
}
````

````python
%%shell
gcc E04_06.c -o E04_06
./E04_06
````

##Example 7
找出滿足餘數條件的最小正數

````python
%%writefile E04_07.c
/*
Find the smallest N that satisfies
    N % x == a
    N % y == b
    N % z == c
For example,
if x=3, y=5, z=7, and a=2, b=3, c=4,
then, N is 53 since
    53 % 3 == 2
    53 % 5 == 3
    53 % 7 == 4
and 53<=3*5*7.

Another test case:
Input:
23 29 37
10 20 30
Output:
14201
 */
#include <stdio.h>
int main(void)
{
    int x, y, z;
    int a, b, c;
    int i;
    scanf("%d%d%d", &x, &y, &z);
    scanf("%d%d%d", &a, &b, &c);
    for (i=1; i<=x*y*z*5; i++) {
          if (i%x==a && i%y==b && i%z==c) {
            printf("%d\n", i);
            break;
          }
    }
    return 0;
}
````

````python
%%shell
gcc E04_07.c -o E04_07
./E04_07
````
